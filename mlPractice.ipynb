```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel('onlineRetail.xlsx')
df.head()
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0, subset=['Invoice'],inplace=True)
df['Invoice']=df['Invoice'].astype('str')
df=df[~df['Invoice'].str.contains('C')]
df
```

```python
basket=(df[df['Country']=="United Kingdom"].groupby(['Invoice','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Invoice'))
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
    
basket_sets=basket.applymap(encode_units)
basket_sets.drop('WHITE METAL LANTERN',inplace=True,axis=1)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets,min_support=0.07,use_colnames=True)
rules= association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules.head()
```

```python
rules[(rules['lift']>=0.6)&(rules['confidence']>=0.8)]
```

```python
frequent_itemsets = apriori(basket_sets,min_support=0.05,use_colnames=True)
rules= association_rules(frequent_itemsets,metric="lift",min_threshold=1)

rules[(rules['lift']>=0.4)&(rules['confidence']>=0.6)]
```

```python
class Employee:

    def __init__(self, empName, salary):
        self.empName = empName
        self.salary = salary

    employeeCount = 0


    def displayEmployee(self):
        print("Name:", self.empName,  ", Salary:", self.salary)

    @classmethod
    def displayCount(cls):
        print("Total Employee:", cls.employeeCount)
        
        
emp1 = Employee("Chishti", 322000)

Employee.employeeCount += 1


emp2 = Employee("Shaheer", 222500)

Employee.employeeCount += 1


emp1.displayEmployee()
emp2.displayEmployee()


Employee.displayCount()
```
